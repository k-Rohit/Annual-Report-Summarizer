```python
import os
import glob

# Folder containing the files
folder_path = "/kaggle/input/annualreports/training/training/gold_summaries"

# Get a list of all the files in the folder
file_pattern = os.path.join(folder_path, "*.txt")
file_list = glob.glob(file_pattern)

# Define a dictionary to group files by their common prefix
file_groups = {}

# Iterate through the file list
for file_path in file_list:
    # Extract the file name without extension
    file_name = os.path.basename(file_path)
    base_name = os.path.splitext(file_name)[0]
    
    # Extract the common prefix (e.g., "33155" from "33155_1.txt")
    common_prefix = base_name.split("_")[0]
    
    # Check if the common prefix exists in the dictionary
    if common_prefix in file_groups:
        # Append the file path to the existing group
        file_groups[common_prefix].append(file_path)
    else:
        # Create a new group with the file path
        file_groups[common_prefix] = [file_path]

# Specify the output directory within the writable Kaggle directory
output_dir = "/kaggle/working/combined_files"
os.makedirs(output_dir, exist_ok=True)

# Iterate through file groups and combine data
for common_prefix, file_paths in file_groups.items():
    # Sort the file paths to maintain the sequence
    file_paths.sort()
    
    # Create the output file path
    output_file_path = os.path.join(output_dir, f"{common_prefix}.txt")
    
    # Combine data from all files in the group
    with open(output_file_path, "w") as output_file:
        for file_path in file_paths:
            with open(file_path, "r") as input_file:
                file_content = input_file.read()
                output_file.write(file_content)

print("Files have been combined and saved in the '/kaggle/working/combined_files' directory.")
```

```python
# Create a zip file containing the combined files
import zipfile
zip_file_path = "/kaggle/working/combined_files.zip"
with zipfile.ZipFile(zip_file_path, "w", zipfile.ZIP_DEFLATED) as zipf:
    for root, _, files in os.walk(output_dir):
        for file in files:
            file_path = os.path.join(root, file)
            arcname = os.path.relpath(file_path, output_dir)
            zipf.write(file_path, arcname)
```
